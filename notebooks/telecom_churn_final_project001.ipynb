Handle Missing Columns Gracefully

If you want to drop columns that may or may not be present in the DataFrame, you can use a list comprehension to drop columns that exist:

```python
#List of columns to drop
columns_to_drop = ['customerID', 'Partner', 'Dependents', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'PaperlessBilling', 'PaymentMethod', 'TotalCharges' ]

#Filter out columns that exist in the DataFrame
columns_to_drop_existing = [col for col in columns_to_drop if col in df.columns]

#Drop the columns that exist
df = df.drop(columns=columns_to_drop_existing)
print("Columns after dropping:", df.columns)
```

```python
#Define features and target
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
#Split the data
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Define the features
numerical_features = ['tenure', 'MonthlyCharges']
categorical_features = ['gender', 'SeniorCitizen', 'PhoneService', 'InternetService', 'StreamingTV', 'StreamingMovies', 'Contract']
```

```python
#Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), #Handle missing values
                                        ('scaler', StandardScaler())])  #Feature scallig

#Preprocessing for categorical data
categorical_transfomer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')), #Handle missing values
                                         ('onehot', OneHotEncoder(handle_unknown='ignore'))]) #onehot encoding


#Combine preprocessing steps into a single ColumnTransformer
preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_features), 
                                               ('cat', categorical_transfomer, categorical_features)])
```

```python
#Define the model
model = RandomForestClassifier(n_estimators=100, random_state=42)

#Create a pipeline that combines preprocessing with the model
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model)])
```

Train the Model

Fit the pipeline to the training data:

```python
pipeline.fit(X_train, y_train)
```

Evaluate the Model

Evaluate the model's performance on the test set:

```python
y_pred = pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

Save the Model

Save the trained model using '.joblib':

```python
joblib.dump(pipeline, 'churn_rates_and_root_causes_project.joblib')
```

Load and Use the Model

To save the saved model later:

```python
#Load the model
loaded_pipeline = joblib.load('churn_rates_and_root_causes_project.joblib')

#Make predictions with the loaded model
y_pred_loaded = loaded_pipeline.predict(X_test)


print("Loaded Model Accuracy:", accuracy_score(y_test, y_pred_loaded))
```
